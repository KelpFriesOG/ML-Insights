# Mergesort

Mergesort is a sorting algorithm that leverages the principles of recursion and a "divide and conquer" strategy to sort an array by performing element-wise comparison in a clever and parallelizable manner.

- **Mergesort does not leverage parallelizability (does not use multiple threads) out of the box, but it can very easily be adapted to be parallelizable.**

---

## Intuition

How many comparisons would it take to sort an array of 2 elements?:

$$[4, 2]$$

A single comparison is enough even in the worst case (the elements are out of order).

How many comparisons would it take to sort an array of 4 elements *if we broke down the array into pairs*?

$$[4, 2, 1, 7]$$

$$[4, 2] [1, 7]$$

First we can sort each subarray with one comparison each:

$$[2, 4] [1, 7]$$

Then lets initialize one pointer in each subarray that starts from the left, we will call them i and j
